# Word Cloud

```python
from scipy.misc import imread
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
text = open("under_sea.txt").read()
```

```python
text = text.lower()
```

```python
stopwords = set(STOPWORDS)
```

```python
wc = WordCloud(max_words=100, stopwords=stopwords, margin=1, random_state=1).generate(text)
```

```python
plt.imshow(wc)
plt.axis("off")
plt.title("20,000 Leagues Under the Sea")
plt.show()
```
